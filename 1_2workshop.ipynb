## Filtering

```python
# filter rows\ instances according to the values of a specific column
# which coutries got more than 10 gold medals
olympics.where(olympics['Gold'] >10).dropna()
```

```python
# filter rows\instances to retain information with a specific column value between 2 values
olympics.where(olympics['Gold'].between(2,10)).dropna()
```

```python
# filter rows\instances to retain information with a specific column containing a specific string
olympics.where(olympics['Country'].str.contains('South Africa')).dropna()
```
