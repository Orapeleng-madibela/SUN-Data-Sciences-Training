**Function with one argument**

```python
#simulate one throw of dice
throw = np.random.randint(1,7)
throw
```

```python
#for the above random number, determine who won. Remo wins if throw = 1, Kele wins if throw equals 2 or 3 and
#Onka wins if throw equals 4, 5, 6
if throw == 1:
  print("Remo wins")
elif throw == 2 or throw == 3:
  print("Kele wins")
else:
  print("Onka wins")

```

```python
#Suppose we want to simulate this 1000 times. We could create a function. Let's first create a function
def winner(a):
  if a==1:
    ans ='Remo'
  elif a==2 or a==3:
    ans = 'Kele'
  else:
    ans = 'Onka'

  return ans
```

```python
#let's run the function for a dice outcome of 5
winner(5)
```

```python
#let's run the function for a dice outcome of 3
winner(3)
```

**Functions with more than one argument**

```python
#array with names of participants
participants = ['Remo','Kele','Onka']
```

```python
#create an function with two inputs.
def winner(a, names):
  if a == 1:
    ans = names[0]
  elif a==2 or a==3:
    ans = names[1]
  else:
    ans = names[2]

  return ans
```

```python
#throw a dice and see who wins by calling the function
throw = np.random.randint(1,7)
[throw,winner(throw,participants)]
```

**Function with two outputs**

```python
#Let's create an function with two inputs and also two outputs
def winner(a,names):
  if a==1:
    ans = names[0]
  elif a==2 or a==3:
    ans = names[1]
  else:
    ans = names[2]

  return [a,ans]
```

```python
#let's run the above function
throw = np.random.randint(1,7)
winner(throw,participants)
```

## Exercise

Suppose you toss a coin and throw a dice. If the coin lands on heads and the dice is less than or equal to 3, Kanshu wins. If the coin lands on heads and dice is greater than 3, Sunday win. If the coin lands on tails and the dice is equal to 1, Sunday wins. If the coin lands on tails and the dice is greater than or equal to 2, Marike win.

Run the above simulation 1000 times and count the number of times that each person wins. Using functions to minimize code.

```python
coin = np.random.choice(['heads','tails'])
dice = np.random.randint(1,7)
[coin,dice]
```

```python
def winner(a,b):
  if a == 'heads':
    if b <= 3:
      ans = 'Kanshu'
    else:
      ans = 'Sunday'
  else:
    if b == 1:
      ans = 'Sunday'
    else:
      ans = 'Marike'

  return[a,b,ans]
```

```python
results = []

for i in np.arange(1000):
  coin = np.random.choice(['heads','tails'])
  dice = np.random.randint(1,7)
  temp = winner(coin,dice)

  results.append(temp[2])

[results.count('Marike'),results.count('Sunday'),results.count('Kanshu')]
```
